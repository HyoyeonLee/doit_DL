## numpy 
   데이터는 배열로 입력하는데 데이터가 커질수록 저차원의 언어를 쓰는 numpy가 적합하다.
   - python의 list &emsp;: 비효율적 (여러 type 요소 수용)    

```python
list1 = [10,'hi',20]      ;  print("* list1 = ", list1);   
list2 = [[1,2,3],[4,5,6]];  print("* list2 = ", list2);
```

- numpy의 array : &emsp;효율적(한 type 요소만 가능)

```python
import numpy as np
from matplotlib import pyplot as plt
arr2 = np.array([[1,2,3],[4,5,6]]);  print("* arr2 =  \n", arr2);
print("type of arr2 : ",type(arr2)); print("sum of elements in arr2 = ",np.sum(arr2));
```

## Matplotlib
과학계의 표준그래프 패키지.
- line vs. scatter plot 

```python
from matplotlib import pyplot as plt
x = np.array([1,2,3,4,5]) ; y = x**2
fig = plt.figure();fig.patch.set_facecolor('xkcd:mint green')
plotLine    = plt.plot(x,y,c='k',label='line');
plotScatter = plt.scatter(x,y,c='r',label='scatter')
plt.xlabel("x",fontsize='large',fontweight='bold');
plt.ylabel("y",fontsize='large',fontweight='bold');
plt.text(3, 20, '$y=x^2$',fontsize=15,bbox={'facecolor':'red', 'alpha': 0.5, 'pad': 10})
plt.legend();plt.show()
```

* random number generater 이용

```python
randX=np.random.randn(1000)
randY=np.random.randn(1000)
fig = plt.figure();fig.patch.set_facecolor('xkcd:mint green')
plt.scatter(randX,randY,s=2)
plt.xlabel("x",fontsize='large',fontweight='bold');
plt.ylabel("y",fontsize='large',fontweight='bold');
plt.text(1, -2, 'Random Number',fontsize=15,bbox={'facecolor':'red', 'alpha': 0.5, 'pad': 10})
plt.show()
```
