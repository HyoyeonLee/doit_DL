## 1-2. 머신러닝 소개
* **<font color="red">머신러닝 Machine Learning</font>** : (기계)학습의 방법 (알고리즘들의 집합)...의미?

|  | Supervised 학습 | Unsupervised 학습 | Reinforcement 학습 |
|:--------------:|----------------------------:|----------------------------:|----------------------------:|
| 정의 | 머신러닝에 주로 쓰이는 학습방식 | Y/N의 타겟없는 경우 | 알파고 같은 것으로 ...|
| 목적 | 데이터분석$\rightarrow$훈련$\rightarrow$규칙찾음 | {$x$}만으로 훈련$\rightarrow$cluster찾음 | agent훈련$\rightarrow$수행$\rightarrow$최대보상|
| 성능평가 | 새 $x$대한 예측값($\hat{y}$)을 타겟($y$)과<br> 비교하여 성능평가 | 타겟($y$)없어서 cluster의 <br>질적평가 어려움 |  |  

&nbsp;  - **데이터**<font color="red">x, y</font> : (훈련/검증/테스트) input data {_**x**_ }와 target _**y**_ 으로 구성 (target 없을 수 도 있어).  
&nbsp;  - **&emsp;모델** <font color="red">y=wx+b</font>:  수식 for (훈련)데이터 ({_**x**_ }와  _**y**_ )로 학습된 머신러닝 알고리즘. (ex) _**y**=w**x**+b_  
&nbsp;  - **&emsp;규칙** <font color="red">w, b</font>: 수식(모델)의 파라미터들 - *w* (weight), *b* (intercept)  
&nbsp;  - **&emsp;훈련** <font color="red">w', b'</font>: 컴퓨터가 데이터의 {_**x**_ }와  _**y**_ 로 모델 파라미터(규칙) 수정  
&nbsp;  - **&emsp;학습** <font color="red">diff(loss)=0</font>: <u>*훈련*</u> 을 통해서 *스스로 규칙찾는*  것. (즉, final, <u>optimized*</u> 파라미터들)      
<font size="3"><font color="green"> &nbsp;&ensp;! 데이터분석하는 훈련자체는 그리 어렵지 않다. Challenges are... <br /> 
    &nbsp; &emsp;&emsp; (1) 다양, 방대한 데이터 확보 <br />
    &nbsp; &emsp;&emsp; (2) 안좋은 데이터 스크리닝  <br /> 
    &nbsp;&ensp;! optimized 의미? Loss Funtion (손실함수,loss)를 minimize하는 방법(최적화 알고리즘) 사용  <br />
    &nbsp; &emsp;&emsp;  :loss 정의는 용도따라 다르다. (ex) loss=1-|y-$\hat{y}$| &emsp;$\hat{y}$=wx+b)</font>
